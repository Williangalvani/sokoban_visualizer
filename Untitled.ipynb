```python
import copy

class Mapa:
    x = 0
    y = 0
    
    _mapa = [[0 for i in range(5)] for i in range(5)]
    _mapa[0][0] = 1
    _mapa[2][2]= 9
    _mapa[3][0]= 9
    _mapa[4][4]= 9
    _mapa[0][3]= 2
    _mapa[0][4]= 2
    _mapa[2][4]= 2
    _mapa[3][2]= 2
    _mapa[4][0]= 2

    moves = []
    
    def step(self):
        try:
            self._mapa[self.y][self.x] = 1
        except:
            print("attempted bad move!")
        return self._mapa
    
    def down(self):
        print("down")
        self.y+=1

    def up(self):
        print("up")
        self.y-=1

    def left(self):
        print("left")
        self.x-=1

    def right(self):
        print("right")
        self.x+=1

    def print(self):
        curmap = copy.deepcopy(self._mapa)
        try:
            curmap[self.y][self.x] = 7
        except:
            pass
        
        print(self.x,self.y)
        for y in range(5):
            print(curmap[y])
        print("")
        
    def __init__(self):
        self.moves = [self.left, self.right, self.up, self.down]
    
mapa = Mapa()
```
